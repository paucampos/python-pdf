## Generando archivo PDF del presupuesto

Instalar librería fpdf

```python
! pip install fpdf
```

Importar la librería

```python
from fpdf import FPDF
pdf = FPDF()

pdf.add_page()
pdf.set_font("Arial", size=12)
pdf.image("template.png", 0, 0, w=200)

pdf.cell(200, 10, txt="Presupuesto del proyecto", ln=True, align='C')
pdf.cell(200, 10, txt="Nombre del proyecto: " + proyecto, ln=True, align='C')
pdf.cell(200, 10, txt="Horas estimadas: " + horas_estimadas, ln=True, align='C')
pdf.cell(200, 10, txt="Valor por hora: " + valor_hora, ln=True, align='C')
pdf.cell(200, 10, txt="Tiempo estimado: " + tiempo_estimado, ln=True, align='C')
pdf.cell(200, 10, txt="Valor total: " + str(total_valor), ln=True, align='C')

pdf.output("presupuesto.pdf")
print("El presupuesto ha sido guardado con éxito en el archivo presupuesto.pdf")
```
